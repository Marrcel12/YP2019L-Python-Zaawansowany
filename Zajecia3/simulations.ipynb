```python
import math
import numpy as np

def Euler(f, xa, xb, ya, n):
    h = (xb - xa) / float(n)
    x = np.zeros(n)
    y = np.zeros(n)
    x[0] = xa
    y[0] = ya
    
    for i in range(1, n):
        y[i] = y[i-1] + h * f(x[i-1], y[i-1])
        x[i] = x[i-1] + h
    return x, y
```

```python
import matplotlib.pyplot as plt
```

```python
x,y = Euler(lambda x, y: math.cos(x) + math.sin(y), 0, 2*3.14, 1, 1000)
```

```python
plt.plot(x,y)
plt.show()
```

```python
def Euler2(f, xa, xb, ya, y1a, n):
    h = (xb - xa) / float(n)
    x = np.zeros(n)
    y = np.zeros(n)
    y1 = np.zeros(n)
    x[0] = xa
    y[0] = ya
    y1[0] = y1a
    for i in range(1, n):
        y1[i] = y1[i-1] + h * f(x[i-1], y[i-1], y1[i-1])
        y[i] = y[i-1] + h * y1[i]
        x[i] = x[i-1] + h
    return x, y
```

```python
x,y = Euler2(lambda x, y, y1: math.sin(x * y) - y1, 0, 10, 1, 1, 1000)
plt.plot(x,y)
plt.show()
```
